```python
# REAL-WORLD PROJECT 1: Student Grading System
print("="*60)
print("PROJECT 1: STUDENT GRADING SYSTEM")
print("="*60)

# Student information (using variables)
student_name = "Rohit Sharma"
student_rollno = 101

# Marks in different subjects (using integers )
maths_marks = 80
science_marks = 89
english_marks = 90
history_marks = 80
computer_marks = 95

# Calculate total and average (using arithmetic operators)
total_marks = maths_marks + science_marks + english_marks + history_marks + computer_marks
average_marks = total_marks / 5
percentage = average_marks  # Since each subject is out of 100

# Display student info (using f-strings)
print(f"Student Name: {student_name}")
print(f"Roll Number: {student_rollno}")
print("-" * 60)

# Display marks (using variables)
print("SUBJECT-WISE MARKS:")
print(f"Mathematics: {maths_marks}")
print(f"Science: {science_marks}")
print(f"English: {english_marks}")
print(f"History: {history_marks}")
print(f"Computer: {computer_marks}")
print("-" * 60)

# Calculate and display totals (using operators)
print(f"Total marks: {total_marks} / 500")
print(f"Percentage: {percentage:.2f}%")

# Determine grade using if-elif-else
print("-" * 60)
if percentage >= 90:
  print("grade: A+")
  print("Remark: Outstanding Performance!")
elif percentage >= 80:
  print("grade: A")
  print("Remark: Excellent Work!")
elif percentage >= 70:
  print("grade: B")
  print("Remark: Good Job")
elif percentage >= 60:
  print("grade: c")
  print("Remark: Satisfactory")
elif percentage >= 40:
  print("grade: D")
  print("Remark: Need Improvement")
else:
  print("grade: Fail")
  print("Remark: Failed- Work Harder")

# Check pass/fail using if-else
print("-" * 60)
if percentage >= 40:
  print("Result: PASSED")

   # Check for distinction using nested if
  if percentage >= 75:
    print("STATUS: PASSED WITH DISTINCTION")
  elif percentage >= 60:
    print("STATUS: PASSED WITH FIRST CLASS")
  else:
    print("STATUS: PASSED")

   # Check scholarship eligibility (using logical operators)
  if percentage >= 85 and maths_marks >= 80 and science_marks >= 80:
    print("SCHOLARSHIP: ✓ ELIGIBLE")
    print("Congratulations! You qualify for Merit Scholarship")
  else:
    print("SCHOLARSHIP: ✗ NOT ELIGIBLE")
else:
  print("RESULT: ✗ FAILED")
  print("Please work harder and try again")

# Subject-wise analysis
print("-"*60)
print("SUBJECT-WISE ANALYSIS:")

if maths_marks >= 80:
  print("Mathematics: EXCELLENT")
elif maths_marks >=40:
  print("Mathematics: PASS")
else:
   print("Mathematics: FAIL - Need to improve")

if science_marks >= 80:
  print("Science: EXCELLENT")
elif maths_marks >=40:
  print("Science: PASS")
else:
   print("Science: FAIL - Need to improve")

if english_marks >= 80:
  print("English: EXCELLENT")
elif maths_marks >=40:
  print("English: PASS")
else:
   print("English: FAIL - Need to improve")

if history_marks >= 80:
  print("History: EXCELLENT")
elif maths_marks >=40:
  print("History: PASS")
else:
   print("History: FAIL - Need to improve")

if computer_marks >= 80:
  print("Computer: EXCELLENT")
elif maths_marks >=40:
  print("Computer: PASS")
else:
   print("Computer: FAIL - Need to improve")
print("-"*60)
```
