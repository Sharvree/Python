```python
# Student Report Card System
import csv

def calculate_garde(marks):
    total = sum(marks)
    percent = total / len(marks)
    if percent >= 90:
        grade = "A"
    elif percent >= 75:
        grade = "B"
    elif percent >= 50:
        grade = "c"
    else:
        grade = "D"
    return total, percent, grade

def save_report(student, marks):
    try:
        total, percent, grade = calculate_garde(marks)
        with open("report_card.csv", "a", newline="")as file:
            writer= csv.writer(file)
            writer.writerow([student, marks, total, percent, grade])
            print("Report saved successful!")
    except Exception as e:
        print("Error saving report:",e)        
               
```

```python
save_report("Sharvree", [85, 90, 78, 92])
```
