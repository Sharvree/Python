```python
# Shopping Card System

 # Available products (tuple list)
products = [("Apple", 50), ("Banana", 20), ("Milk", 40)]

# Shopping cart dictionary
card = {}

def add_items(item, qty):
  for name, price in products:
    if name == item:
      if item in card:
        card[item]["qty"] += qty
      else:
         card[item] = {"price": price, "qty": qty}
      return
      print("Item not found")

def remove_item(item):
  if item in card:
    del card[item]
  else:
     print("Item not in card!")

def view_card():
    print("Your Card:")
    for item, details in card.items():
       print(f"{item} - {details['qty']} x {details['price']} = {details['qty']*details['price']}")

def checkout():
    total = sum(details["qty"] * details["price"] for details in card.values())
    print(f"Total Bill: ₹{total}")

# Example usage
add_items("Apple", 2)
add_items("Milk", 1)
remove_item("Apple")
view_card()
checkout()
```
